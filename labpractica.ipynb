```python
#python program to add BMI columen in data frame

#BMI caluclated as   = BMI = weight/(height) 2


import pandas as pd

#Create sample dataset

data = {
    'name ': ['A','B','C','D'],
    'Weight' : [60,70,80,90],
    'height': [1.6,1.7,1.8,175]
}

df = pd.DataFrame(data)

#Caluclate BMI

df['BMI'] = df['Weight'] /(df['height'] **2)

#Display resutl
print(df)
```

```python
#pyhton program to get number of observation , missing values and NaN Values

import pandas as pd

#load dataset
df = pd.read_csv("iris.csv")

# Number of observationS(rows)
num_obs = df.shape[0]
print("number of observation" , num_obs)

#Missing values(column -wise)
missing_values = df.isnull().sum()
print("Missing valules in coluumn wise: ", missing_values)

#total NaN values
total_nan = df.isnull().sum().sum()
print("Total NaN values:", total_nan)
```
