```python
#python program to add BMI columen in data frame

#BMI caluclated as   = BMI = weight/(height) 2


import pandas as pd

#Create sample dataset

data = {
    'name ': ['A','B','C','D'],
    'Weight' : [60,70,80,90],
    'height': [1.6,1.7,1.8,175]
}

df = pd.DataFrame(data)

#Caluclate BMI

df['BMI'] = df['Weight'] /(df['height'] **2)

#Display resutl
print(df)
```

```python
#python program to draw scatter plots to compare two features of iris dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load datset
df = pd.read_csv("iris.csv")

#Scatter plot : sepal length Vs sepal width
sns.scatterplot(x = 'sepal_length' , y ='sepal_width', hue = 'species',data= df)
plt.title("sepal length vs sepal width")
plt.xlabel("sepal length")
plt.ylabel("Sepal width")
plt.show()

#scatter plot:  petal length vs petal width
sns.scatterplot(x = "petal_length" , y = "petal_width", hue ='species', data= df)
plt.title("Petal length vs petal width")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()


```

```python
# py prog to find relationships between petal length and petal width

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load datasete
df = pd.read_csv("iris.csv")

#scatter plt to show realtionship
sns.scatterplot(x="petal_length", y="petal_width", hue='species' ,data=df)

#label and title
plt.title("Realtionships between petal length and petal widht")
plt.xlabel("petal length")
plt.ylable("petal width")
plt.show()
```

```python
#py prog to create histogram for the three species of iris dataset 

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("iris.csv")

# Separate data for each species
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']

# Histogram for Petal Length of each species
plt.hist(setosa['petal_length'], alpha=0.5, label='Setosa')
plt.hist(versicolor['petal_length'], alpha=0.5, label='Versicolor')
plt.hist(virginica['petal_length'], alpha=0.5, label='Virginica')

# Labels and Title
plt.title("Histogram of Petal Length for Iris Species")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.legend()

# Show plot
plt.show()
```
